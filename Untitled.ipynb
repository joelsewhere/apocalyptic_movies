```python
import pandas as pd
import json

with open('titles.json') as fp:
    data = json.load(fp)
```

```python
df = pd.DataFrame(data)
```

```python
import string
import re
```

```python
punctuation_letters = string.punctuation + string.ascii_letters
```

```python
def strip_date(text):
    stripped = text.translate(str.maketrans(' ', ' ', punctuation_letters))
    stripped = stripped.rstrip()
    stripped = stripped.replace('â€“', ' ')
    
    return stripped
```

```python
test = df.copy()
```

```python
test.date = test.date.apply(strip_date)
```

```python
def select_first_date(text):
    split = text.split()
    return split[0]
```

```python
test.date = test.date.apply(select_first_date)
```

```python
test = test.reset_index(drop = True)
```

```python
test.date = pd.to_datetime(test.date)
```
